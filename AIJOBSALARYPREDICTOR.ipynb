```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, MultiLabelBinarizer
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from category_encoders import TargetEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['job_title','experience_level','employment_type','company_location','company_size','employee_residence','required_skills','education_required','industry','company_name']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['job_title','experience_level','employment_type','company_location','company_size','employee_residence','required_skills','education_required','industry','company_name']),

                                       remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores = []
for i in range(200):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(x_train,y_train)
  y_pred = pipe.predict(x_test)
  scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```
